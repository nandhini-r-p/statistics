***Z-score*** - Z-score gives how much std is the value (i) away from mean

```python
outliers=[]

def detect_outliers(dataset):

  # outliers are any values after 3rd standard deviation
  threshold = 3
  mean = np.mean(dataset)
  std = np.std(dataset)

  for i in dataset:
    z_score = (i-mean) / std

    if np.abs(z_score) > threshold:
      outliers.append(i)

  return outliers
```

```python
detect_outliers(dataset)
```

***IQR*** - Inter Quartile Range

Steps to find IQR:
1. Sort the data
2. Calculate Q1 and Q3
3. IQR = Q3-Q1
4. Lower fence = Q1 - 1.5(IQR)
5. Upper fence = Q3 + 1.5(IQR)

```python
dataset = sorted(dataset)
dataset
```

```python
q1, q3 = np.percentile(dataset, [25, 75])
q1, q3
```

```python
iqr = q3 - q1
```

```python
# find lower and upper fence

lower_fence = q1 - (1.5*iqr)
upper_fence = q3 + (1.5*iqr)

lower_fence, upper_fence
```
