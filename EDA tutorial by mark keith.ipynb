EDA is the process of analyzing data sets to summarize their characteristics using the combination of statistical calculations and data visualizations.

**Univariate Analysis**

There are 3 types of univariate statistics that we are interested in for each var:
* General Info: data type, count of total values, number of unique values
* Range and middle: min, max, mean, median, mode, quartiles
* Normality & spread: Std, skewness, kurtosis


**Normality**

In normality, we want to understand the shape, spread of each variable. We refer to this as the "normality" of each variable. A variable has the normal distribution(Gaussian, Gauss or Laplace-Gauss) when there is bell curve formed when plotted on a graph. Here we will find how normal our variables are
